```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from icecream import ic
```

```python
# import data from ../data/data-agg.csv 
data = pd.read_csv('../data/data-agg.csv')

# fill the table with the other relevant values 
data['N_throws'] = data['N_start_heads_up'] + data['N_start_tails_up']
data['tails_tails'] = data['N_start_tails_up'] - data['tails_heads']
data['heads_tails'] = data['N_start_heads_up'] - data['heads_heads']

data['falls_on_tails'] = data['tails_tails'] + data['heads_tails']
data['falls_on_heads'] = data['tails_heads'] + data['heads_heads']
data['falls_on_tails_ratio'] = data['falls_on_tails'] / data['N_throws']
data['falls_on_heads_ratio'] = data['falls_on_heads'] / data['N_throws']


print(data)
```

```python
# Group by 'person' and compute the mean of 'falls_on_tails'
means_tails_df = data.groupby("person")[["falls_on_tails_ratio"]].mean()

# Compute the number of unique persons
num_persons = data["person"].unique()

print(f"Number of unique persons: {len(num_persons)}")

plt.hist(means_tails_df["falls_on_tails_ratio"], bins=20, edgecolor='black')
plt.xlabel("Falls on Tails per Throw Ratio")
plt.ylabel("Tails per throw")
plt.show()

# Group by 'person' and compute the mean of 'falls_on_heads'
mean_heads_df = data.groupby("person")[["falls_on_heads_ratio"]].mean()

# Compute the number of unique persons
num_persons = data["person"].unique()

print(f"Number of unique persons: {len(num_persons)}")

plt.hist(mean_heads_df["falls_on_heads_ratio"], bins=20, edgecolor='black')
plt.xlabel("Falls on Tails per Throw Ratio")
plt.ylabel("Tails per throw")
plt.show()
```
